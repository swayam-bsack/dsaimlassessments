```python
'''1. Load the 'tips' dataset from Seaborn and display the first 5 rows.  '''
import seaborn as sns
import matplotlib.pyplot as plt
# Load the 'tips' dataset
df=sns.load_dataset("tips")
df.head(5)
```

```python
'''2. Create a scatter plot of total_bill vs tip using the 'tips' dataset. '''
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.title('Scatter Plot of Total Bill vs Tip')
```

```python
'''3. Plot a line graph for total_bill vs tip using Seaborn'''
sns.lineplot(data=df, x='total_bill',y='tip')
```

```python
'''4.Draw a bar plot to show the average tip for each day in the 'tips' dataset. '''
sns.barplot(data=df, x="day", y="tip", estimator='mean')
#estimator is used to calculate the mean of the tips for each day
#possible values for estimator are mean, median, sum, etc.
plt.show()
```

```python
'''5. Use a count plot to display how many records are available for each day.'''
sns.countplot(x="day", data=df)
```

```python
'''6. Make a box plot to show total_bill distribution across different days.'''
sns.boxplot(x="total_bill", y="day", data=df)
plt.title('Box Plot of Total Bill Distribution Across Days')
```

```python
'''7. Create a violin plot for total_bill vs day.'''
sns.violinplot(x="total_bill",y="day",data=df)
```

```python
'''8. Plot a histogram of the 'total_bill' column with KDE using histplot. '''
sns.histplot(x="total_bill",bins=15,kde=True,data=df)
plt.grid()
```

```python
'''9. Load the 'iris' dataset and draw a pairplot for all features.'''
df1 = sns.load_dataset("iris")
df1.head(5)
#checking different values of species
sns.pairplot(data=df1, hue='species')
```

```python
'''10. Use the 'titanic' dataset and show a heatmap of its numeric correlation matrix.'''
df2=sns.load_dataset("titanic")
df2.head(5)
corr=df2.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm",annot=True)
```

```python
'''11. Use the 'penguins' dataset and draw a scatter plot of bill_length_mm vs 
flipper_length_mm'''
df3=sns.load_dataset("penguins")
sns.scatterplot(data=df3, x='bill_length_mm', y='flipper_length_mm')
```

```python
'''12. Plot a box plot for 'fare' against 'class' using the 'titanic' dataset. '''
sns.boxplot(x="class",y="fare",data=df2)
#boxplot is used to show the distribution of fare across different classes in the titanic dataset
```

```python
'''19. Apply a dark theme to your plots using set_style and recreate a scatter plot. '''
sns.set_style("darkgrid")#set_style options: whitegrid, darkgrid, white, dark, ticks
sns.scatterplot(data=df, x='total_bill', y='tip')
```

```python
'''20. Convert all categorical columns in the 'tips' dataset to numeric and create a full 
correlation heatmap. '''
import pandas as pd
# Convert categorical columns to numeric using pd.get_dummies
#.get_dummies() is used to convert categorical variable(s) into dummy/indicator variables
df_numeric = pd.get_dummies(data=df, drop_first=True)
#drop_first=True is used to avoid the dummy variable trap by dropping the first category of each categorical variable
#df_numeric.head(5)
#df.head(5)
# Create a full correlation heatmap
corr_full = df_numeric.corr(numeric_only=True)
sns.heatmap(corr_full, cmap="coolwarm", annot=True)
```
