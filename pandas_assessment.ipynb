```python
# 1.Create a Pandas Series using a list of 5 fruits. Print the 3rd fruit using index. 
import pandas as pd
l=["Apple","Banana","Melon","Orange","Grapes"]
s=pd.Series(l)
print(s[2])
```

```python
#2. You are given a list of student names and their scores. Convert it into a Series and 
#display the score of the student named "Anu".
l2=["Anu", "Bala", "Chitra", "Deepak"]
scores=[85, 90, 78, 88]
s=pd.Series(scores, index=l2)
print(s["Anu"])
```

```python
#3.Create a DataFrame with columns Name, Age, Marks. Display only the Marks column.
d={
    "Name":["Swayam","Ram","John","Dean"],
    "Age":[19,18,18,18],
    "Marks":[100,50,45,67]
}
df=pd.DataFrame(d)
print(df["Marks"])
```

```python
#4. From a DataFrame of employees with columns Name, Salary, and Department, use .loc 
#to display the salary of the employee named "Raj".
d={
    "Name":["Swayam","Raj","John","Dean"],
    "Salary":[19000,1875,18,18],
    "Department":["DS","Ai","ML","OS"]
}
df=pd.DataFrame(d)
print(df.loc[df["Name"]=="Raj","Salary"].values[0])  # .values converts to a numpy array and [0] gets the first element
#print(df.loc[1, "Salary"])  # Using index to access Raj's salary directly
```

```python
#Q5. Use .iloc to print the first 3 rows from a DataFrame containing student information.
d={
    "Name":["Swayam","Ram","John","Dean"],
    "Age":[19,18,18,18],
    "Marks":[100,50,45,67]
}
df=pd.DataFrame(d)
print(df.iloc[0:3])
#print(df.iloc[0:3,0])- would display only the names
```

```python
#Q6. Given a DataFrame of customer records, use .at to update the value of a specific cell at 
#row index 2 and column Name.
d={
    "Name":["Swayam","Ram","John","Dean"],
    "Age":[19,18,18,18],
    "Salary":[100,50,45,67]
}
df=pd.DataFrame(d)
df.at[2,"Name"]="John Cena"
df
```

```python
#Q7. Create a DataFrame with some NaN values in the Marks column. Replace all NaN values 
#with the average (mean) of the Marks column. 
data={
    "Name":["Swayam","Ram","John","Dean"],
    "Age":[19,18,18,18],
    "Marks":[100,50,None,67]  # None represents NaN
}
df=pd.DataFrame(data)
mean_marks=df["Marks"].mean()
df["Marks"]=df["Marks"].fillna(mean_marks)
#df["Marks"].fillna(mean_marks,inplace=True)-Not recommended
df
```

```python
#Q14. Create a DataFrame from a dictionary of items (Item, Quantity, Price) and calculate the 
#total price for each row using a new column Total = Quantity * Price.
data = {
    "Item": ["Apple", "Banana", "Orange"],
    "Quantity": [10, 20, 15],
    "Price": [1.5, 0.5, 2.0]
}
df1=pd.DataFrame(data)
df1["Total"]=df1["Quantity"]*df1["Price"]
df1
```

```python
'''Q15. You have a DataFrame of city temperatures. Some values in the Temperature column 
are missing. Fill them using the average temperature calculated using mean(). '''
data = {
    "City": ["New York", "Los Angeles", "Chicago", "Houston"],
    "Temperature": [30, None, 25, 28]
}
df2=pd.DataFrame(data)
df2["Temperature"]=df2["Temperature"].fillna(df2["Temperature"].mean())
df2
```
