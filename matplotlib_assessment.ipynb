```python
from matplotlib import pyplot as plt
```

```python
'''1. Plot a simple line graph using the values: x = [1, 2, 3, 4], y = [10, 20, 25, 30]  
2. Add a title and axis labels to the line graph in Question 1. '''
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x,y)
plt.title('Simple Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
'''3. Plot a red dotted line for: x = [0, 1, 2, 3], y = [3, 6, 1, 8] '''
x=[0,1,2,3]
y=[3,1,6,8]
plt.plot(x,y,linestyle="dotted",color="red")
```

```python
'''4. Plot two lines in the same graph: x = [1, 2, 3, 4], y1 = [10, 20, 25, 30], y2 = [5, 15, 20, 25]  
5. Add a legend to the above plot to differentiate the two lines. '''
x=[1, 2, 3, 4]
y1=[10, 20, 25, 30]
y2=[5, 15, 20, 25]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
```

```python
'''6. Plot a bar chart for: Fruits = ['Apple', 'Banana', 'Mango'], Quantity = [10, 20, 15]  
7. Plot a horizontal bar chart for the same fruit and quantity data. '''
Fruits = ['Apple', 'Banana', 'Mango']
Quantity = [10, 20, 15]
plt.bar(Fruits,Quantity)
```

```python
plt.barh(Fruits,Quantity)
plt.xlabel("Quantity")
plt.ylabel("Fruits")
```

```python
'''8. Create a pie chart with percentage labels: Subjects = ['Math', 'Science', 'English'], Values = 
[30, 40, 30] '''
choose_colors=["red","pink","black"]
Subjects = ['Math', 'Science', 'English']
Values = [30, 40, 30]
plt.pie(Values, labels=Subjects, autopct='%1.1f%%', colors=choose_colors)#%1.1f%%- 1.- min width of the percentage,.1-one decimal place, %%- % sign
plt.show()
```

```python
'''9. Create a scatter plot: x = [5, 7, 8, 7], y = [8, 5, 6, 3] 
10. Create a histogram for the data: marks = [60, 70, 80, 75, 70, 90, 85, 80]'''
x = [5, 7, 8, 7]
y = [8, 5, 6, 3]
plt.scatter(x, y)
```

```python
marks = [60, 70, 80, 75, 70, 90, 85, 80]
plt.hist(marks, bins=5, edgecolor='black')# bins=5- number of bins, edgecolor='black'- color of the edges of the bars
```

```python
'''11. Create a subplot with two plots: one line plot and one bar chart. 
12. Change the figure size to 8x5 while plotting a simple graph. 
13. Plot a graph with customized line style, marker, and color. 
14. Save a plot as a PNG file using plt.savefig() 
15. Add grid lines to any plot you create.'''

# Set figure size to 8x5 (Number 12)
plt.figure(figsize=(8, 5))#plt.figure(figsize=(width, height)) - width- width of the figure, height- height of the figure

# First subplot: Line Plot (Number 13)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y, linestyle='--', marker='o', color='blue')  # Customized line style, marker, and color
plt.grid(axis='both')  # Add grid lines (Number 15)
plt.title('Line Plot')

# Second subplot: Bar Chart (Number 11)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
subjects = ['Math', 'Science', 'English']
values = [30, 40, 30]
plt.bar(subjects, values, color='green')  # Bar chart
plt.grid(axis='y')  # Add grid lines to y-axis (Number 15)
plt.title('Bar Chart')

'''Subplot concept:
subplot(nrows, ncols, index) - nrows- number of rows, ncols- number of columns, index- index of the plot (1-based indexing)
subplot(1, 2, 1) - 1 row, 2 columns, first plot within the overall figure
subplot(1, 2, 2) - 1 row, 2 columns, second plot within the overall figure

'''
# Save the figure as a PNG file (Number 14)
plt.savefig("plot.png")

# Show plots
plt.show()
```

```python
'''16. Plot a line graph and annotate the highest point on it.'''
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
max_y=max(y)
max_x=x[y.index(max_y)]
plt.annotate(f'Highest Point: ({max_x}, {max_y})',xy=(max_x,max_y))  # Annotate the highest point
#plt.annotate(text, xy, xytext=None, arrowprops=None); text and xy are required parameters,
#xy- point to annotate, xytext- text position, arrowprops- properties of the arrow
plt.title('Line Graph with Annotation')
plt.show()
```

```python
'''17. Create a bar chart and add value labels on top of each bar.'''
Fruits = ['Apple', 'Banana', 'Mango']
Quantity = [10, 20, 15]
plt.bar(Fruits, Quantity, color='orange')
for i, v in enumerate(Quantity): #enumerate() - returns index and value of the iterable
    plt.text(i, v, v, ha="center",va="bottom")  # plt.text(x, y, s, ha, va) - x and y are coordinates, 
        #s is the text to display, ha is horizontal alignment, va is vertical alignment
        # x,y,s are required parameters, ha and va are optional parameters
'''use of enumerate() function:
for i, v in enumerate(Quantity):
    print(f"Index: {i}, Value: {v}")
'''
```

```python
'''18. Create a plot with x-axis rotated 45 degrees using tick labels: Months = ['Jan', 'Feb', 
'Mar', 'Apr'], Sales = [250, 300, 150, 400]'''
Months = ['Jan', 'Feb', 'Mar', 'Apr']
Sales = [250, 300, 150, 400]
plt.plot(Months, Sales)
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
#plt.yticks(rotation=45)  # Rotate y-axis labels by 45 degrees
plt.grid(axis='y')
plt.show()
```
